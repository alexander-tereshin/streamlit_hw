```python
import os
import pathlib
import requests

import pandas as pd

from sqlalchemy import create_engine
from dotenv import load_dotenv, find_dotenv


load_dotenv(find_dotenv())
```

```python
postgres_ip = os.getenv('POSTGRES_IP')
postgres_pass = os.getenv('POSTGRES_PASS')
```

```python
engine = create_engine(f"postgresql://teal:{postgres_pass}@{postgres_ip}:5432/streamlit_hw_db")

with engine.connect() as conn:
    result = conn.execute("select 'hello from remote postgres on vps!'")
    print(result.first())
```

```python
url = 'https://github.com/aiedu-courses/stepik_linear_models/tree/main/datasets'

response = requests.get(url)
paths = [i['path'] for i in response.json()['payload']['tree']['items']]

for url in paths:
    if len(i:= url.split('/')[1].split('.')[0].split('_')) == 1:
        df_name = i[0].lower()
    else:
        df_name = i[1].lower()
        
        
    locals()[f'df_{df_name}'] = pd.read_csv(f'https://raw.githubusercontent.com/aiedu-courses/stepik_linear_models/main/{url}',
                                            index_col=0)
    print(f'df_{df_name} added to locals')
    
    try:
        locals()[f'df_{df_name}'].to_sql(df_name, engine)
        
        print(f'{df_name} saved in remote postgres\n')
        
    except:
        with engine.connect() as conn:
            result = conn.execute(f"drop table {df_name}")
            
        locals()[f'df_{df_name}'].to_sql(df_name, engine)
        
        print(f'{df_name} saved in remote postgres\n')
```
